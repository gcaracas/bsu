```python
import networkx as nx
import pandas as pd
import  pickle
```

```python
G=nx.read_gexf("graph.bz2")
```

```python
df = pd.DataFrame(columns=['SRC','TGT','VOT','RES','YEA','DAT','TXT','ID'])
```

```python
cnt=0
cnt2 = 0
for elem in G.edges.data():
    cnt2 = cnt2+1
    cnt = cnt+1
    if cnt == 1000:
        print(cnt2)
        cnt = 0
    df = df.append({'SRC':elem[0], 'TGT':elem[1],'VOT':elem[2]["VOT"],'RES':elem[2]["RES"],'YEA':elem[2]["YEA"],'DAT':elem[2]["DAT"],'TXT':elem[2]["TXT"],'ID':elem[2]["id"]}, ignore_index=True)
```
