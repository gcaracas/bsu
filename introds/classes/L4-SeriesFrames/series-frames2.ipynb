# Your Turn - GDP vs Alcohol Consumption
## All data is for year 2010 (the only available table year for alcohol consumption by country on Wikipedia)

```python
alc_by_country = pd.read_csv("alc_by_country.csv")
alc_by_country = pd.DataFrame(index=alc_by_country["Country"],
                             data={'2010 Total': list(alc_by_country['Total']),
                                  'Beer(%)': list(alc_by_country['Beer(%)']),
                                  'Wine(%)': list(alc_by_country['Wine(%)']),
                                  'Spirits(%)': list(alc_by_country['Spirits(%)']),
                                  'Other(%)': list(alc_by_country['Other(%)'])})

gdp_by_country = pd.read_csv("gdp_by_country.csv")
# Rename Country column to country
gdp_by_country = gdp_by_country.rename(columns = {'Country (or dependent territory)':'Country'})
# Only use Country and 2010 entry
gdp_by_country = gdp_by_country[['Country', '2010']]
gdp_by_country = pd.DataFrame(index=gdp_by_country["Country"],
                             data={'2010 GDP (USD)': list(gdp_by_country['2010'])})

# Remove commas from data for more accurate sorting later
gdp_by_country['2010 GDP (USD)'] = gdp_by_country['2010 GDP (USD)'].str.replace(',', '')
# Drop all NaN entries
gdp_by_country = gdp_by_country.dropna()
# Convert from Object type to Int type
gdp_by_country['2010 GDP (USD)'] = gdp_by_country['2010 GDP (USD)'].astype(int)

high_gdp = gdp_by_country['2010 GDP (USD)'] > gdp_by_country['2010 GDP (USD)'].mean()
high_alc = alc_by_country['2010 Total'] > alc_by_country['2010 Total'].mean()

print("Crosstabulation of High GDP and High Consumption:")
print(pd.crosstab(high_gdp, high_alc))
print("\n")

gdp_bin = pd.qcut(gdp_by_country['2010 GDP (USD)'], 3, labels=("Low", "Median", "High"))
alc_bin = pd.qcut(alc_by_country['2010 Total'], 3, labels=("Low", "Median", "High"))
print("GDP and Alcohol Consumption binned and cross-tabulated:")
print(pd.crosstab(gdp_bin, alc_bin))
```

## Analysis

This results seem to indicate that there is a threshhold just below the average GDP that enables an individual to drink more-than-average per capita.  Simply cross-tabulating against mean GDP show that while a large percentage of below-mean GDP nations have higher than mean drinking rates, more have lower-than-mean drinking rates, while almost all higher-than-mean nations have higher-than-mean drinking.

However, this is just by mean and does not take into account the huge disparities that nations like the US cause in the calculation of that mean.

Quantifying and binning the data before cross-tabulating shows that it is instead almost a linear increase in drinking as gdp rises.  High GDP nations have the largest overall high drinking rates, while median GDP nations have a median drinking rate, and low-GDP nations the lowest.  This more or less makes sense, since if you don't have money you can't buy booze.

This also doesn't necessarily factor in higher-than-average GDP nations that have simply outlawed drinking altogether, like Saudi Arabia, and indeed many of the nations that happen to be on the lower-end of the GDP range are small nations who have outlawed it entirely, usually for religious purposes.
